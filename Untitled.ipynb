```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
from sklearn import datasets, linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('SeoulBikeData.csv')
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Date"]
```

```python
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month
df["day"] = df["Date"].dt.day
df
```

```python
df.info()
```

```python
df_catagoric_features = df.select_dtypes(include=object)
df_catagoric_features.columns
```

```python


dummy_encoded_variables = pd.get_dummies(df_catagoric_features, drop_first=True)
df_numeric_features = df.select_dtypes(include=np.number)
```

```python
df = pd.concat([df_numeric_features,dummy_encoded_variables], axis=1)
df.head()
```

```python
plt.figure(figsize = (10, 5))
df['Rented Bike Count'].plot(kind ="hist")
```

```python
plt.figure(figsize = (10, 5))
df['Visibility (10m)'].plot(kind ="hist")
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.pointplot(data=df, x='Temperature(C)', y='Rented Bike Count', hue='year', ax=ax)
ax.set(title='What are the factors that influence the number of bikes rented ')
```

```python
plt.figure(figsize=(9,8))
plt.figure(figsize=(15,10))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.RdYlBu)
plt.show()
```

```python
target_corr = abs(cor['Rented Bike Count'])
positive_corr_target = target_corr[target_corr> (0.2)]
positive_corr_target
```

```python
X = df.drop(['Rented Bike Count'],axis=1).values   
y = df['Rented Bike Count'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.30)
print(x_train.shape, y_train.shape) 
print(x_test.shape, y_test.shape)
```

```python
lrm = linear_model.LinearRegression()
lrm_model = lrm.fit(x_train, y_train)
lrm_predictions = lrm.predict(x_test)
```

```python
lrm_predictions
```

```python
plt.scatter(y_test, lrm_predictions)
plt.title("Rented Bike Count")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
print('score',lrm_model.score(x_test, y_test))
```
